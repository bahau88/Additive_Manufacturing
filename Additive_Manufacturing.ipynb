```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/bahau88/Additive_Manufacturing/main/data.csv')
```

```python
df.info()
```

```python
df.head(10)
```

```python
df['infill_pattern'] = df['infill_pattern'].replace({'grid': 1, 'honeycomb': 2})
df['material'] = df['material'].replace({'abs': 1, 'pla': 2})
df.rename(columns={'tension_strenght': 'tension_strength'}, inplace=True)
df.head(10)
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

```

```python

# Define features and target
features = ['layer_height', 'wall_thickness', 'infill_density', 'infill_pattern', 'nozzle_temperature', 'bed_temperature', 'print_speed', 'material', 'fan_speed',]
target = 'roughness'

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Train Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Evaluate the performance
y_pred = rf.predict(X_test)
print('R2 score:', r2_score(y_test, y_pred))

# Get feature importances from the model
importances = rf.feature_importances_

# Create a bar plot of feature importances using Plotly
colors = ['Coral', 'orange', 'Crimson', 'green', 'DarkCyan', 'purple', 'blue', 'red', 'orange']  # Specify different colors for each bar
fig_rf = go.Figure([go.Bar(x=features, y=importances, text=importances, textposition='auto', textfont=dict(size=12),
                           marker=dict(color=colors)
                           )])

# Update layout and axes properties
fig_rf.update_layout(
    title='Parameter to Roughness',
    xaxis_title='Feature',
    yaxis_title='Importance',
    width=700,  # set width of the plot
    height=300,  # set height of the plot
    font=dict(size=9),
    plot_bgcolor='white',
    xaxis=dict(tickangle=45)
)

fig_rf.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import plotly.graph_objects as go

# Define features and targets
features = ['layer_height', 'wall_thickness', 'infill_density', 'infill_pattern', 'nozzle_temperature', 'bed_temperature', 'print_speed', 'material', 'fan_speed']
targets = ['roughness', 'tension_strength', 'elongation']

# Split data into training and testing sets for each target
X_train, X_test, y_train_roughness, y_test_roughness = train_test_split(df[features], df[targets[0]], test_size=0.2, random_state=42)
X_train, X_test, y_train_tension, y_test_tension = train_test_split(df[features], df[targets[1]], test_size=0.2, random_state=42)
X_train, X_test, y_train_elongation, y_test_elongation = train_test_split(df[features], df[targets[2]], test_size=0.2, random_state=42)

# Train Random Forest Regressors for each target
rf_roughness = RandomForestRegressor(n_estimators=100, random_state=42)
rf_roughness.fit(X_train, y_train_roughness)

rf_tension = RandomForestRegressor(n_estimators=100, random_state=42)
rf_tension.fit(X_train, y_train_tension)

rf_elongation = RandomForestRegressor(n_estimators=100, random_state=42)
rf_elongation.fit(X_train, y_train_elongation)

# Predict for each target
y_pred_roughness = rf_roughness.predict(X_test)
y_pred_tension = rf_tension.predict(X_test)
y_pred_elongation = rf_elongation.predict(X_test)

# Calculate R2 scores for each target
r2_roughness = r2_score(y_test_roughness, y_pred_roughness)
r2_tension = r2_score(y_test_tension, y_pred_tension)
r2_elongation = r2_score(y_test_elongation, y_pred_elongation)

print('R2 score (Roughness):', r2_roughness)
print('R2 score (Tension Strength):', r2_tension)
print('R2 score (Elongation):', r2_elongation)
```

```python
import numpy as np

# Get feature importances for each target
importances_roughness = rf_roughness.feature_importances_
importances_tension = rf_tension.feature_importances_
importances_elongation = rf_elongation.feature_importances_

# Create a heatmap for feature importances
fig = go.Figure()

fig.add_trace(go.Heatmap(z=[importances_roughness, importances_tension, importances_elongation],
                         x=features,
                         y=targets,
                         colorscale='Viridis', zmin=0, zmax=np.max([importances_roughness, importances_tension, importances_elongation])))

fig.update_layout(
    title='Feature Importances for Quality Metrics',
    xaxis_title='Feature',
    yaxis_title='Quality Metric',
    width=900,
    height=500,
    xaxis=dict(tickangle=45),
    plot_bgcolor='white'
)

fig.show()
```

Machine Learning

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

# Define features and targets
features = ['layer_height', 'wall_thickness', 'infill_density', 'infill_pattern', 'nozzle_temperature', 'bed_temperature', 'print_speed', 'material', 'fan_speed']
targets = ['roughness', 'tension_strength', 'elongation']

# Split data into features and targets
X = df[features]
y_roughness = df['roughness']
y_tension_strength = df['tension_strength']
y_elongation = df['elongation']

# Split data into training and testing sets
X_train, X_test, y_train_roughness, y_test_roughness = train_test_split(X, y_roughness, test_size=0.2, random_state=42)
X_train, X_test, y_train_tension_strength, y_test_tension_strength = train_test_split(X, y_tension_strength, test_size=0.2, random_state=42)
X_train, X_test, y_train_elongation, y_test_elongation = train_test_split(X, y_elongation, test_size=0.2, random_state=42)

# Train Random Forest Regressors for each target
rf_roughness = RandomForestRegressor(n_estimators=100, random_state=42)
rf_roughness.fit(X_train, y_train_roughness)

rf_tension_strength = RandomForestRegressor(n_estimators=100, random_state=42)
rf_tension_strength.fit(X_train, y_train_tension_strength)

rf_elongation = RandomForestRegressor(n_estimators=100, random_state=42)
rf_elongation.fit(X_train, y_train_elongation)

# Predict quality metrics for the test set
predicted_roughness = rf_roughness.predict(X_test)
predicted_tension_strength = rf_tension_strength.predict(X_test)
predicted_elongation = rf_elongation.predict(X_test)

# Calculate R2 scores for each prediction
r2_roughness = r2_score(y_test_roughness, predicted_roughness)
r2_tension_strength = r2_score(y_test_tension_strength, predicted_tension_strength)
r2_elongation = r2_score(y_test_elongation, predicted_elongation)

print('R2 score (Roughness):', r2_roughness)
print('R2 score (Tension Strength):', r2_tension_strength)
print('R2 score (Elongation):', r2_elongation)

# Calculate MAE for each prediction
mae_roughness = mean_absolute_error(y_test_roughness, predicted_roughness)
mae_tension_strength = mean_absolute_error(y_test_tension_strength, predicted_tension_strength)
mae_elongation = mean_absolute_error(y_test_elongation, predicted_elongation)

print('Mean Absolute Error (Roughness):', mae_roughness)
print('Mean Absolute Error (Tension Strength):', mae_tension_strength)
print('Mean Absolute Error (Elongation):', mae_elongation)

# Example: Predict quality metrics for a new set of parameters
new_parameters = [[0.02, 7, 90, 2, 225, 65, 40, 1, 25]]
predicted_roughness_new = rf_roughness.predict(new_parameters)
predicted_tension_strength_new = rf_tension_strength.predict(new_parameters)
predicted_elongation_new = rf_elongation.predict(new_parameters)

print('Predicted Roughness:', predicted_roughness_new[0])
print('Predicted Tension Strength:', predicted_tension_strength_new[0])
print('Predicted Elongation:', predicted_elongation_new[0])
```
